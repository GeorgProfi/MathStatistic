```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Генерация выборок
np.random.seed(42) 
n = 50
x = np.linspace(0, 1, n)
y = 2 * x + 1 + 0.1 * np.random.randn(n)  #Пример линейной связи

# Построение графиков
plt.figure(figsize=(12, 5))

# График значений переменных
plt.subplot(1, 3, 1)
plt.plot(x, y, 'o')
plt.title('График значений переменных')
plt.xlabel('x')
plt.ylabel('y')

# График облака пар точек
plt.subplot(1, 3, 2)
plt.scatter(x, y)
plt.title('Облако пар точек')
plt.xlabel('x')
plt.ylabel('y')

# Проверка наличия связи
correlation, p_value = pearsonr(x, y)
plt.subplot(1, 3, 3)
plt.text(0.5, 0.5, f'Корреляция: {correlation:.2f}\np-значение: {p_value:.4f}', horizontalalignment='center', verticalalignment='center', fontsize=12)
plt.axis('off')
plt.title('Проверка наличия связи')
plt.tight_layout()
plt.show()
if p_value < 0.05:
    print('Существует статистически значимая связь между переменными.')
else:
    print('Статистически значимой связи между переменными не выявлено.')
```
